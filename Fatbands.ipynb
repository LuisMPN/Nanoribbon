```python
import sisl
import sisl.viz
import matplotlib.pyplot as plt
```

```python
H = sisl.get_sile("nanoribbon.fdf").read_hamiltonian()
```

```python
band_struct = sisl.BandStructure(H, points=[[0, 0, 0], [1/2, 0, 0]],
    divisions=100, names=[r"\Gamma", "K"]
)
# Then we can plot the bands
band_struct.plot()
```

```python
# Get the fatbands plot
fatbands = band_struct.plot.fatbands(Erange=[-2,2])
# Split the contributions by the n and l quantum numbers
#fatbands.split_groups(on="n+l",scale=10)
fatbands.update_settings(groups=[
    {"atoms":0, "spin":0, "name":"Atom 0 spin 0", "color": "blue", "scale":"5"},
    {"atoms":0, "spin":1, "name":"Atom 0 spin 1", "color": "red", "scale":"5"},
    {"atoms":3, "spin":0, "name":"Atom 3 spin 0", "color": "blue", "scale":"5"},
    {"atoms":3, "spin":1, "name":"Atom 3 spin 1", "color": "red", "scale":"5"}])
```
